# 1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import time
import numpy as np
from bs4 import BeautifulSoup
import requests
start_url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
page_response = requests.get(start_url)
page_response.status_code # code 200 means connection was sucessful and it is now active
soup = BeautifulSoup(page_response.content, 'html.parser')

soup = BeautifulSoup(page_response.content, 'html.parser')
container=[]
for i in soup.find_all('tbody'):
    container.append(i.text.split("\n"))

rank=[]
for i in range(14,247,8):
    rank.append(container[1][i])

name=[]
for i in range(15,250,8):
    name.append(container[1][i].split("[")[0])

artist=[]
for i in range(16,251,8):
    artist.append(container[1][i])

views=[]
for i in range(17,251,8):
    views.append(container[1][i])

upload_date=[]
for i in range(18,254,8):
    upload_date.append(container[1][i])
zipped= list(zip(rank,name,artist,views,upload_date))

df= pd.DataFrame(zipped, columns=['Rank','Video','Artist','Views(in Billion)','Uploaded On'])
    
df.to_csv('wikipedia_youtube_video.csv')

data=pd.read_csv('wikipedia_youtube_video.csv')

data.drop('Unnamed: 0',axis=1)
```

# 2. Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:
A) Match title (I.e. 1
st ODI)
B) Series
C) Place
D) Date
E) Time
Note: - From bcci.tv home page you have reach to the international fixture page through code

```python
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome(r"D:\chromedriver.exe")  # Optional argument, if not specified will search path.

driver.get('https://www.bcci.tv/.');

time.sleep(5) # Let the user actually see something!
```

```python
wait = WebDriverWait(driver, 10)
element = wait.until(EC.element_to_be_clickable((By.XPATH, '/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a')))
```

```python
element.click()
```

```python
match_titles=[]
series=[]
place=[]
date=[]
Time=[]
```

```python
container=[]
Product_Name_tag=driver.find_elements(By.XPATH,"//div[@class='fixture-card-top']")
for k in Product_Name_tag:
    p=k.text.split("\n")
    container.append(p)
```

```python
bucket=[]
Name_tag=driver.find_elements(By.XPATH,"//div[@class='fixture-card-bottom']")
for k in Name_tag:
    n=k.text.split('-')
    bucket.append(n)
```

```python
for i in container:
    series.append(i[0])
    date.append(i[1])
    Time.append(i[2])
```

```python
for i in bucket:
    match_titles.append(i[0])
    place.append(i[1])
```

```python
for i in range(len(place)):
    a=place[i].split("\n")
    place[i]=a[0]
```

```python
driver.quit()
    
zipped = list(zip(match_titles,series,place,date,Time))
    
df = pd.DataFrame(zipped, columns=['Match Number', 'Series', 'Place','Date','Time'])
        
for col in df.columns:
    df[col] = df[col].fillna('-')
    
df.to_csv('Indian_team_matches.csv')

data=pd.read_csv("Indian_team_matches.csv")
data.drop("Unnamed: 0",axis=1,inplace=True)
data
```

```python
ratings=[]
for i in range(0,len(rating),23):
    ratings.append(rating[i])
```

```python
data=pd.DataFrame({'Name':name,"Run Time":run,"Genre":Type,'Year Span':year,'Ratings':ratings,})
    
data=data[0:100]
    
data.to_csv('imdb.csv')                         
data=pd.read_csv("imdb.csv")
data
```
